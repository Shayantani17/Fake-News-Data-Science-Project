```python
import pandas as pd

# Load the metrics from CSV files
logistic_regression_metrics = pd.read_csv('logistic_regression_metrics.csv')
random_forest_metrics = pd.read_csv('random_forest_metrics.csv')
naive_bayes_metrics = pd.read_csv('naive_bayes_metrics.csv')
svm_metrics = pd.read_csv('svm_metrics.csv')
knn_metrics = pd.read_csv('knn_metrics.csv')

# Add a column to identify the model type
logistic_regression_metrics['Model'] = 'Logistic Regression'
random_forest_metrics['Model'] = 'Random Forest'
naive_bayes_metrics['Model'] = 'Naive Bayes'
svm_metrics['Model'] = 'SVM'
knn_metrics['Model'] = 'KNN'

# Combine all metrics into a single DataFrame
all_metrics = pd.concat([
    logistic_regression_metrics,
    random_forest_metrics,
    naive_bayes_metrics,
    svm_metrics,
    knn_metrics
], ignore_index=True)

# Display the combined DataFrame in table form
print(all_metrics.to_markdown())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Melt the DataFrame to long format for easier plotting with seaborn
metrics_melted = all_metrics.melt(id_vars=["Model"], var_name="Metric", value_name="Score")

# Create a bar plot
plt.figure(figsize=(16, 12))
bar_plot = sns.barplot(x="Model", y="Score", hue="Metric", data=metrics_melted, palette="muted")

# Add the values on top of the bars
for p in bar_plot.patches:
    bar_plot.annotate(format(p.get_height(), '.2f'), 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='center', 
                      xytext=(0, 20), 
                      textcoords='offset points')

plt.title('Performance Based on Accuracy, Precision and F1 Scores')
plt.ylabel('Score Values')
plt.xlabel('Classifiers')
plt.ylim(0, 140)  # Assuming the metrics are scaled between 0 and 1
plt.legend(loc='upper right')
plt.show()
```
